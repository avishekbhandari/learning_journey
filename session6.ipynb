Syntax

UPDATE - modifying existing data, used to update existing record in a table.

Syntax

Primary key - column used to identify records uniquely, establish connection between tables

candidate key - any column that can uniquely identify a record, null is not allowed, more than one in a table

Alternate Key- keys that are not selected as primary key, null is not allowed

Composite Key - key made of two or more columns (primary key made of two or more columns)

Unique key - ensures all values in a column are unique, can have null value,can have multiple unique keys for eg. email

super key - set of columns that uniquely identifies a record

surrogate key - an artificial unique identifier or system generated (basically primary key)

Referential Integrity - ensure that relationship between tables remain consistent and valid enforced using FK
                    - A FK value in one table must always a valid PK value in another table

Referential Integrity Error - occurs when try to insert, update or delete the data that breaks the relationship between tables

CASCADE - delete both parent and child at same time so that referntial integrity does not occur

DROP vs DELETE

DROP - deletes the entire table structure plus the table's data permanently, DDL command
DELETE - only deletes data(rows), removes data and table still exist, can use where condition as well, DML command

Delete means removing student from classroom, but classroom still exist
DROP means destroying the entire classroom and nothing remains

we use DELETE when we want to remove data but keep the table, but DROP remove whole table

TRUNCATE vs DELETE

Truncate -> removes all rows instantly(no where clause allowed)
->It deletes all rows at once, reset table to empty state and faster than DELETE
->Free up the space which is occupied by the table
->No rollback is possible

TRUNCATE TABLE emp; #all data removed instantly but table structure remains

INSERT INTO emp VALUES(4,"Dack",4000); # Autoincrement reset is possible in TRUNCATE

When should you use TRUNCATE instead of DELETE?
-->when we want to remove all data quickly and when there is no need to filter rows and performance is important


SQL command are mainly divided into 5 categories
1. DDL (Data Definition Language)
    -> used to create and modify dbs structure
    -> CREATE, ALTER, DROP, TRUNCATE, RENAME
   
3. DML (Data Manipulation Language)
   ->used to work with data inside tables (kind of modifying table datas)
   ->INSERT, UPDATE, DELETE

   
3. DQL (Data Query Language)
   -> used to fetch data
   -> SELECT

   
4. DCL (Data Control Language)
   -> used for permissions and access control
   -> GRANT, REVOKE
   -> GRANT SELECT ON emp TO u1;
   -> REVOKE SELECT ON emp FROM u1;

   
5. TCL (Transaction Control Language)
    ->used to manage transactions (group of queries)
   -> COMMIT -save changes, ROLLBACK - cancel transaction, SAVEPOINT
