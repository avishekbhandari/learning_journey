```python
#printing only even number using for loop
data = [1,2,3,4,5,6,8,12,16,13,99]

result = []

for d in data:
    if d%2 == 0:
        result.append(d)
print(result)
```

```python
#printing only even number using function and for loop

def even(x):
    return x%2 == 0  #returns True if number is even otherwise False

result = []
    
for d in data:
    if even(d): #when this condition is true it appends value to list/select values that satisfy even-number condition
        result.append(d)
print(result)
```

```python
#printing only even number using filter 

def even(x):
    return x%2 == 0
   
result = filter(even, data)
print(list(result))
```

```python
#printing only even number using filter and lambda

result = filter(lambda x : x%2 == 0, data)
print(list(result))

```
