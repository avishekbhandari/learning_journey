Built-in servers: 2 main servers

    1. NameNode
    2. DataNode
    
    Manage and store data in a cluster.


HDFS architecture

    Master-slave architecture

        NameNode
        -> master server/master of HDFS cluster
        -> manage file system/knows where the file is stored
        -> control client access
        -> file operation

        DataNode
        -> worker node
        -> every machine in the cluster typically has 1 DataNode
        -> store actual data
        -> r and w blocks
        -> manage block replication
        -> block creation
        -> block deletion
        -> block replication

        Block
        -> data is stored in blocks
        -> size 128 MB
        -> smallest unit of data that HDFs can read and write
        -> each block will be stored in different machine
    

Hadoop moves computation to where data is stored

MapReduce
    -> processing framework in hadoop
    -> process huge dataset
    -> run tasks in parallel
    -> use cluster of commodity hardware
    -> written in Java

    Flow:
        Map phase
            Input → key-value pairs
        Shuffle & Sort
            Groups by key
        Reduce phase
            Aggregates results

    Map Function
        -> process input data
        -> converts into key-value pairs
        -> breaks large data into smaller pieces

        Input: "apple banana apple"
        Map: (apple,1), (banana,1), (apple,1)
        Reduce: (apple,2), (banana,1)
