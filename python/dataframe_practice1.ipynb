```python
import pandas as pd

data = {
    "Product": ["Laptop","Keyboard","Mouse","SSD","Phone"],
    "Price": [80000,5000,2000,10000,50000],
    "Sales": [40,100,10,20,60]
    
}

df = pd.DataFrame(data)
print(df)
```

```python
#adding column

df["Revenue"] = df["Sales"] * df["Price"]
print(df)
```

```python
#printing first three rows

print(df.head(3))
```

```python
#printing datatypes

print(df.dtypes)
```

```python
#printing datatypes

print(df.info())
```

```python
#printing single column datatype

print(df["Sales"].dtype)
```

```python
#adding new record using loc

df.loc[len(df)] = ["Monitor", 75000, 10, 750000]
print(df)
```

```python
#adding missing values

df.loc[len(df)] = ["Ipad", 40000, 20, None]
print(df)
```

```python
#detecting null values
print(df.isnull())
```

```python
#counting missing values

print(df.isnull().sum())
```

```python
#filling null values

df["Revenue"].fillna(0, inplace=True)
print(df)
```

```python
#displaying products where sales is less than 30 and sort by aescending order
low_sales= df[df["Sales"] < 30].sort_values(by="Revenue", ascending=True)
print(low_sales)
```

```python
#recompute

df["Revenue"] = df["Sales"] * df["Price"]
print(df)
```

```python
#Which product generates the highest revenue?

highest_revenue = df.loc[df["Revenue"].idxmax()]  # .loc - used to access rows by label or index, idmax()- index of maximum value
print(highest_revenue)
```

```python
#Which product has the lowest sales but relatively high price?

#finding minimum sales value in the datset
lowest_sales = df["Sales"].min()
#print(lowest_sales)

#filtering rows which has minimum sales
lowest_sales_df = df[df["Sales"] ==lowest_sales]
#print(lowest_sales_df)

#among filtered rows selecting the product which has highest revenue
lowest_sales_hp = lowest_sales_df.loc[lowest_sales_df["Revenue"].idxmax()]
print(lowest_sales_hp)
```

```python
#Which product is most dependent on volume (high sales, low price)?

high_sales = df["Sales"].max()
#print(high_sales)

high_sales_df = df[df["Sales"] ==high_sales]
#print(high_sales_df)

high_sales_lp = high_sales_df.loc[high_sales_df["Price"].idxmin()]
print(high_sales_lp)
```
