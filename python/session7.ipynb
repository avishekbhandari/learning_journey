```python
# create data frame and add column to it

import pandas as pd

data = {
    "product": ["laptop","phone","tablet","watch","airpods"],
    "price": [1000, 2000, 4000, 5000,6000],
    "sales": [10,20,30,50,40]
}

#df =pd.DataFrame(data, index=["product1","product2","product3","product4","product5"])
df =pd.DataFrame(data)
print(df)
```

```python
# add column

df["revenue"] = df["price"] * df["sales"]
print(df)
```

```python
# add a record to the data
# add new record using LOC - to locate row
#to return one or more specified rows

df.loc[len(df)] = ["macbook", 100000, 25, 2500000]
print(df)
```

```python
#calculate total revenue

total_revenue = df["revenue"].sum()
print(total_revenue)
```

```python
top_product = df.loc[df["revenue"].idxmax()] # .loc - used to access rows by label or index, idmax()- index of maximum value
print(top_product)
```

```python
#add new record using LOC
# loc[len(df)] - adds data at the next index position
df.loc[len(df)] = ["book", 700000, 10, 7000000] #len(df) - returns number of rows
print(df)
```

```python
# to see index value 0 and 1

print(df.loc[[0,1]])
```

```python
#missing values

df.loc[2,"sales"]=None
print(df)
```

```python
#filling missing values
df["sales"].fillna(0, inplace=True) #fillna() ->replace all NaN values with 0
print(df)
```

```python
# Filter high-value products based on price threshold

#df["price"] - selects price column and return a series
#df["price"] > 13000 - creates a boolean mask
#df[df["price"] > 13000] - filters row when condition = True called boolean indexing

high_value = df[df["price"] > 13000]
print(high_value)
```

```python
#sort column

sorted_df = df.sort_values(by="revenue", ascending=False)
print(sorted_df)
```

```python
import pandas as pd

data = {
    "product": ["laptop","phone","tablet","watch","airpods"],
    "price": [40000, 2000, 4000, 5000,6000],
    "sales": [None,20,30,50,40]
}


df =pd.DataFrame(data)
print(df)
```

```python
#detect missing values - isnull()
#isnull() returns True if value is missing and false if value is present
print(df.isnull())
```

```python
#count missing values
print(df.isnull().sum())
```

```python
#filling missing values
#fillna(0) -replaces any value with 0 or any value you passing, inplace=True - updates the original dataframe
df.fillna(0, inplace=True)
print(df)
```

```python
import pandas as pd

data = {
    "product": ["laptop","phone","tablet","watch","airpods"],
    "price": [40000, 2000, 4000, 5000,6000],
    "sales": [None,20,30,50,40]
}


df =pd.DataFrame(data)

#filling missing values with mean average

df["sales"].fillna(df["sales"].mean(), inplace=True)
print(df)
```

```python
#remove missing values
df = df.dropna()
print(df)
```
