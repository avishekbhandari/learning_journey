filter -builtin python function used to select elements from a dataset/iterable(list,tuple) based on a condition
->it keeps only those values where the condition is True
->returns iterator (not list)
apply condition -> keep valid data -> remove unwanted data

filter(function(condition - return True/False), iterable(list/dataset))
-> function used in filter must return True of False
map -> transforms data/updates data
filter() -> select subset of data
lambda() -> define small/inline logic

```python
#get only even number

#using for loop

num = [1,2,3,4,5,6]

res4 = []

for n in num:
    if n%2 == 0:
        res4.append(n)
print(res4)
```

```python
# using filter 
res5 = filter(lambda x: x%2 ==0, num) #filter return an iterator
print(list(res5))
```

```python
#get employees with salary gretar than 10000

s = [5000, 10000, 8000, 11000, 14000]

high_sal = filter(lambda x : x > 10000, s)
print(list(high_sal))
```

Recursion -> when a function calls itself
base case -> condition that stops recursion
Recursive case -> function calls itself with a modified input to approach the base case

```python
#print numbers from 1 to 5
def print_numbers(n):
    if n==6:  #base case
        return  #exit the function immediately, do not run further code
    print(n)
    print_numbers(n+1) #recursive case


print_numbers(1)    
```

fibonacci series

->sequence where 0,1,1,2,3,5,8,13
->each number is the sum of the previous two (f(n) = f(n-1) + f(n-2))
-> use recursion

```python
def fib(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fib(n-1) + fib(n-2)

print(fib(5))
```

```python
# if(1)=1, f(0)=0
```

```python
def fib(n):
    if n<=1:
        return n
    return fib(n-1) + fib(n-2)

#print series
x = 7

for i in range(x):
    print(fib(i))
```
