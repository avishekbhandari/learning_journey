zip function - combines multiple collections into pairs, it groups element based on index position

```python
coders = ["nisha", "hari", "avishek", "bipina", "manoj"]
marks = [50,60,70,80,90]
list(zip(coders,marks))
```

```python
# index wise mapping 
# index 0 -  nisha + 50
# zip returns zip object not a list
```

```python
zip(coders,marks)
```

```python
alphabet = ["A", "B", "C"]
num = [80,90]
list(zip(alphabet,num))
```
