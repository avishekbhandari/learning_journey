Data warehousing

-> process of collecting, cleaning and storing data
-> Data warehouse is the place where structure/processed data is stored for analysis and reporting

Hadoop
-> open-source distributed computing framework
-> store and process large-scale datasets across a cluster of commodity hardware
-> use multiple computers together instead of just one
-> distributed environment (Breaks data into small pieces and distributes those pieces across many computers and process them in parallel)
-> fault tolerance
-> cost-effective

HDFS (Hadoop Distributed File System)



Clusters
-> group of computers connected together that work as a single system
-> Hadoop runs on clusters.


How Hadoop Works

->divides data in to blocks
->stores blocks across multiple machine
->processes block in parallel using MapReduce
->Combines result in final output
