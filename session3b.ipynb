```python
# passing arguments
def myfunction(f_name):  #f_name is the parameter
    print(f_name + " Avi")
myfunction("you") # "you" is the argument
myfunction("pp")
myfunction("cc")
```

```python
# parameters vs arguments
#parameter: it is the variable inside the () in the function definition
#argument: it is the actual value that is send to the function when called
```
