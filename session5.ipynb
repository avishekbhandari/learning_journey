OOP - Encapsulation
->restricting direct access to data
->controlling access using methods (get/set)
->prevents direct modification of internal state
->data hiding,control access
->protects data/security
->prevents accidental changes
->make code modular
->Public,Protected and Private - Access modifiers

```python
#public access modifier - no restriction, direct access allowed

class Employee:
    def __init__(self,name,salary):
        self.name = name
        self.salary = salary
        
emp = Employee("Saroj", 10000)

print(emp.name)
print(emp.salary)     
```

```python
#protected access modifier - python does not strictly restrict its a warning for developers

class employee:
    def __init__(self,name,salary):
        self._salary=salary

emp = employee("Avi", 10000)
print(emp._salary)
```

```python
# Real encapsulation

class employee:
    def __init__(self,name,salary):
        self._salary=salary

    def get_salary(self):
        return self._salary

emp = employee("Avi", 10000)
print(emp.get_salary())
```

```python
#private access modifier 
# - strong restriction
# - uses name mangling (__var -> __className__Var)
#python internally changes - _employee__salary

class employee: # define a class named Employee
    def __init__(self,name,salary):    # constructor method (runs automatically when object is created)
        self.name=name    #public attribute(self.name) - attributes belongs to an object - name(variable/parameter) 
        self.__salary=salary    #private attribute (name mangling, internally becomes _employee__salary)

emp = employee("Avi", 10000)   # create object of Employee class
#print(emp.__salary) #gives attribute error

print(emp.name)    #accessing public variable
print(emp._employee__salary)    # accessing private variable using name mangling 
```

```python
# Real encapsulation
class employee:
    def __init__(self,name,salary):
        self.__salary=salary

    def get_salary(self):
        return self.__salary

emp = employee("Avi", 10000)

print(emp.get_salary())
```

```python
class employee1:
    def __init__(self,salary):
        self.__salary = salary #private attribute

    #Setter method - update salary safely
    def set_salary(self, new_salary):
        if new_salary > 0:
            self.__salary = new_salary

    #Getter method - used to access private data safely
    def get_salary(self):
        return self.__salary

emp = employee1(10000) #employee1.__init__(emp,10000)
emp.set_salary(20000) #internally - employee1.set_salary(emp,20000)
print(emp.get_salary()) #employee1.get_salary(emp)
```

Inheritance
-> allows class to inherit attributes and members from another class
-> promotes code reusability, avoid duplication, easy maintenance, extensibility
-> one class(Child) can use properties and methods of another class(Parent).

```python
#Inheritance with super

class Person:
    def __init__(self,name):
        self.name = name

    def show_name(self):
        print("Name: ", self.name)

class Employee(Person): #inheriting Person
    def __init__(self,name,salary):
        super().__init__(name)  #calling parent constructor
        self.salary = salary
       # self.name = name
    #Super call method - calls method of the Parent class, to initalize the parent properties, to reuse parent logic
    #it allows to call method from Parent class without explicitly naming it

emp = Employee("Avi", 20000)
#emp.show_name()  #inherited method
print(emp.salary)
print(emp.name)   
    
```

```python
#Inheritance without super

class Person:
    def __init__(self,name):
        self.name = name

    def show_name(self):
        print("Name: ", self.name)

class Employee(Person): #inheriting Person
     def __init__(self,name,salary):
         self.salary = salary
         self.name = name

emp = Employee("Avi", 20000)
emp.show_name() 
print(emp.salary)

    
```

key difference with and without super()

Without Super() - parent not initialized, Missing Data
With Super() - Parent initialized, Complete Object, Correct Approach

inheritance - reuse code from Parent
super() - calling parent's setup

Polymorphism
-> same method name, different behaviour
-> allows method to do different things based on object calling them
-> Method Overriding(runtime Polymorphism) 
-> Method OverLoading

```python
# Method Overriding(runtime Polymorphism) - when child class provides its own implementation  of a method defined in parent class

class Parent:
    def show(self):
        print("This is a person")

class Employee(Parent):
    def show(self): #overriding parent Method
        super().show() #call parent method - Parent Logic + child logic
        print("This is an employee")
        
obj = Employee()
obj.show()

#python checks -> child class then -> Parent Class - child method overrides parent method
```

```python
# Method Overloaing
# same function, handle different inputs

def add(a,b = 0 , c = 0):
    return a+b+c

#same function behaves differently based on inputs
print(add(10))
print(add(10,20))
print(add(10,20,30))
```

```python
#args

def add(*numbers):
    return sum(numbers)

print(add(10))
print(add(10,20))
print(add(10,20,30))
```

Abstraction
-> hiding complex implementation details and showing only the necessary features
-> reduces complexity, improves readability, hide unnecessary details
-> make code more secure, helps in large systems
-> we have Abstract Classes and Abstract Methods
-> Abstract Class - a class that cannot be instantiated
-> Abstract Method - a method that is declared but not implemented
