```python
# constructor(__init__)
# all classes have buit-in methods called __init__
# always executed when the class is called
# used to assign values to object properties
```

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

s1 = Person("avishek","21")
print(s1.name)
print(s1.age)
```

```python
# without __init__ - you need to set properties manually for each object
class Person1:
    pass

p1 = Person1()
p1.name = "Nisha"
p1.age = 18

print(p1.name)
print(p1.age)
```

```python
#using __init__ - you set initial values when creating the object
class Person2:
    def __init__(self,name,age):
        self.name = name
        self.age = age
p3 = Person2("avi", 25)
print(p3.name)
print(p3.age)
```
