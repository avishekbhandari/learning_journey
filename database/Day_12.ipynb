### Having Clause
->Having after grouping

```python
%sql
--find department with average salary greater than 50000
SELECT department, avg(salary) as avrg_salary FROM emp1 GROUP BY department HAVING avrg_salary > 50000;
```

```python
%sql

-- find department with more than 2 employees
SELECT department, count(*) as num_employees FROM emp1 GROUP BY department HAVING num_employees > 2;
```

```python
%sql

--find departments where average salary is greater than 50000 only include IT and Finance departments
SELECT department, AVG(salary) as avg_salary FROM employees WHERE department IN ('IT', 'Finance') GROUP BY department HAVING avg_salary > 50000;
```


LIKE - Pattern Match
% - Any character

BETWEEN - Range
UPDATE - Modify data

### INNER JOIN
-> only matching records from tables

```python
%sql

--get employees with their department names only matching records

SELECT e.emp_id, e.emp_name, d.department_name FROM employee e INNER JOIN department d ON e.department_id = d.department_id;
```

#### LEFT JOIN
-> all records from left table and only matching record from right table

```python
%sql

--get all employees even if they donot have department
SELECT e.emp_id, e.emp_name, d.department_name FROM employee e LEFT JOIN department d ON e.department_id = d.department_id;
```

```python
%sql


SELECT e.emp_id, e.emp_name, d.department_name FROM employee e RIGHT JOIN department d ON e.department_id = d.department_id;
```

```python
%sql
SELECT e.emp_id, e.emp_name, d.department_name FROM employee e FULL OUTER JOIN department d ON e.department_id = d.department_id;
```

####CROSS JOIN

```python
%sql
SELECT e.emp_id, e.emp_name, d.department_name FROM employee e CROSS JOIN department d;
```
