##count difference
###explain difference between COUNT(*), COUNT(1) and  COUNT(COLUMN)

COUNT(*) - COUNT ALL ROWS, INCLUDE NULL
COUNT(1) - COUNT ROWS, SAME AS COUNT(*)
COUNT(column) - counts non-null values, ignores null


###find total salary per department

```python
%sql

SELECT dept, sum(salary) as total_salary FROM emp GROUP BY dept;
```

##counting employee in each department

```python
%sql

SELECT dept,count(*) as emp_count FROM emp GROUP BY dept;
```


##find must recently hired employee date

```python
%sql

SELECT max(hire_date) AS latest_hire_date FROM emp;
```


##find must recently hired employee data

```python
%sql

SELECT * FROM emp WHERE hire_date ==(SELECT max(hire_date) FROM emp);
```

### find highest salary per department

```python
%sql
SELECT dept,max(salary) as max_salary FROM emp GROUP BY dept ORDER BY max_salary DESC;
```

### find avg salary per department

```python
%sql

SELECT dept,avg(salary) as avg_salary FROM emp GROUP BY dept;
    
```

```python
###count employees which is hired after 2000
```

```python
%sql

SELECT count(*) from emp where hire_date > '2000-01-01';
    
```

```python
%sql
select *,
row_number() over(partition by department order by salary desc) as rank
from employees;
```

###select the employee who are top 3 as per salary in each department

```python
%sql
select *
from(
  select *,
  row_number() over(partition by department order by salary desc) as rank
  from employees
)temp
where rank <= 3;

--assigns rank to each employee based on their salary in each department. The top 3 employees in each department are selected.
```
